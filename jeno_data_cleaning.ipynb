```python
import pandas as pd
import numpy as np
import seaborn as sb

df = pd.read_csv('dataset_halfSecondWindow.csv')
```

```python
fig = sb.heatmap(df.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
columns = ['activityrecognition#1', 'android.sensor.accelerometer#mean',
       'android.sensor.game_rotation_vector#mean',
       'android.sensor.gravity#mean', 'android.sensor.gyroscope#mean',
       'android.sensor.gyroscope_uncalibrated#mean',
       'android.sensor.linear_acceleration#mean',
       'android.sensor.magnetic_field#mean',
       'android.sensor.magnetic_field_uncalibrated#mean',
       'android.sensor.orientation#mean',
       'android.sensor.rotation_vector#mean', 'sound#mean', 'speed#mean', 'target', 'user']
df_features = df[columns]
fig = sb.heatmap(df_features.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
df_features.sort_values(by='user')
```

```python
df_copy = df_features
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer

ord_encoder = OrdinalEncoder(categories=[df_copy['target'].unique(), df_copy['user'].unique()])
df_copy[['target', 'user']] = ord_encoder.fit_transform(df_copy[['target', 'user']])
```

```python
df_features[['target', 'user']] = df_copy[['target', 'user']].astype(int)
df_copy[['target', 'user']] = df_copy[['target', 'user']].astype(int)
```

```python
df_copy = df_copy.groupby(['user', 'target']).transform(lambda x: x.fillna(x.mean()))
fig = sb.heatmap(df_copy.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
df_copy[['target', 'user']] = df_features[['target', 'user']]
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df_copy = imputer.fit_transform(df_copy)
```

```python
df_copy_final = pd.DataFrame(df_copy, columns=df_features.columns)
df_copy_final[['target', 'user']] = df_copy_final[['target', 'user']].astype(int)
df_copy_final
```

```python
df_copy_final.user.value_counts()
```

```python
df_target = df_copy_final[(df_copy_final["user"].isin([3, 0, 7, 4]))]

df_target.to_csv('target.csv')
```

```python
df_test = df_copy_final[(df_copy_final["user"].isin([1, 2, 5, 6, 8, 9, 10, 11, 12]))]
df_test.to_csv('test.csv')
```
