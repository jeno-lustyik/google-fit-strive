```python
import pandas as pd
import numpy as np
import seaborn as sb

df = pd.read_csv('dataset_halfSecondWindow.csv')
```

```python
df.isna().sum()
```

```python
df
```

```python
fig = sb.heatmap(df.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
fig_cor = sb.heatmap(df.corr(), cmap="YlGnBu", cbar_kws={'label': 'Correlation'})
```

```python
columns = ['activityrecognition#1', 'android.sensor.accelerometer#mean',
       'android.sensor.game_rotation_vector#mean',
       'android.sensor.gravity#mean', 'android.sensor.gyroscope#mean',
       'android.sensor.gyroscope_uncalibrated#mean',
       'android.sensor.linear_acceleration#mean',
       'android.sensor.magnetic_field#mean',
       'android.sensor.magnetic_field_uncalibrated#mean',
       'android.sensor.orientation#mean',
       'android.sensor.rotation_vector#mean', 'sound#mean', 'speed#mean', 'target', 'user']
df_features = df[columns]
fig = sb.heatmap(df_features.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
fig_cor = sb.heatmap(df_features.corr(), cmap="YlGnBu", cbar_kws={'label': 'Correlation'})
```

```python
df_features.sort_values(by='user')
```

```python
df_features['user'] = df_features['user'].apply(lambda row: int(row.replace('U', '')))
df_features
```

```python
df_features.groupby(['user', 'target'])['sound#mean'].get_group((1, 'Bus')).mean()
```

```python
df_copy = df_features
df_features
```

```python
df_copy = df_copy.groupby(['user', 'target']).transform(lambda x: x.fillna(x.mean()))
```

```python
df_copy[['user', 'target']] = df_features[['user', 'target']]
fig = sb.heatmap(df_copy.isna(), cmap="YlGnBu", cbar_kws={'label': 'Missing Data'})
```

```python
df_copy_final = df_copy.dropna()
df_copy_final.to_csv('40k_rows.csv')
```
